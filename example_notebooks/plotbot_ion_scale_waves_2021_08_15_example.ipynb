   - Possible bidirectional L-mode waves (Doppler-shifted)

## üìä Appendix Event Analysis - August 26, 2019

Let's analyze the sampling cadence and frequency mapping for the appendix event from the paper.


```python
# Define appendix event time range
trange_appendix = ['2019/08/26 19:00:00', '2019/08/26 20:50:00']

print(f"Appendix event time range: {trange_appendix}")
print("This event shows dual-polarization wave packets from 19:00 to 20:50 UT")
```

```python
# Load appendix event data and check sampling cadence
print("Loading appendix event data to check sampling cadence...")

# Get the magnetic field data for the appendix event
mag_data_appendix = get_data(trange_appendix, 'mag_rtn')

# Check the time stamps to determine sampling cadence
if hasattr(mag_data_appendix, 'time') and len(mag_data_appendix.time) > 1:
    # Calculate time differences
    time_diffs = np.diff(mag_data_appendix.time)
    dt_median = np.median(time_diffs).total_seconds()
    sampling_rate = 1.0 / dt_median
    
    print(f"\nüìä Sampling Cadence Analysis:")
    print(f"   Number of data points: {len(mag_data_appendix.time)}")
    print(f"   Median time step: {dt_median:.6f} seconds")
    print(f"   Sampling rate: {sampling_rate:.2f} samples/second")
    print(f"   Nyquist frequency: {sampling_rate/2:.2f} Hz")
    
    # Store for frequency mapping calculation
    original_sampling_rate = sampling_rate
else:
    print("Could not determine sampling cadence from data")
    # Use typical PSP/FIELDS sampling rate as fallback
    original_sampling_rate = 292.97  # samples per second
    print(f"Using typical PSP/FIELDS sampling rate: {original_sampling_rate:.2f} samples/second")
```

```python
# Calculate frequency mapping for 100kHz audio playback
audio_sampling_rate = 100000  # 100 kHz as specified

# Calculate time duration of appendix event
from datetime import datetime
start_time = datetime.strptime('2019/08/26 19:00:00', '%Y/%m/%d %H:%M:%S')
end_time = datetime.strptime('2019/08/26 20:50:00', '%Y/%m/%d %H:%M:%S')
event_duration_seconds = (end_time - start_time).total_seconds()

print(f"\nüî¢ Frequency Mapping Calculation:")
print(f"   Event duration: {event_duration_seconds:.0f} seconds ({event_duration_seconds/60:.1f} minutes)")
print(f"   Audio sampling rate: {audio_sampling_rate:,} Hz")
print(f"   Original data sampling rate: {original_sampling_rate:.2f} Hz")

# Calculate time compression ratio
# When audifying, the original time series gets mapped to audio frequencies
# The time compression ratio determines how original frequencies map to audio frequencies
time_compression_ratio = original_sampling_rate / audio_sampling_rate

print(f"   Time compression ratio: {time_compression_ratio:.8f}")
print(f"   This means original frequencies are multiplied by {1/time_compression_ratio:.0f} in the audio")
```

```python
# Create frequency mapping chart
print(f"\nüìà Frequency Mapping Chart (100kHz Audio Playback)")
print(f"=" * 60)
print(f"{'Audio Frequency (Hz)':>20} | {'Original Frequency (Hz)':>25}")
print(f"=" * 60)

# Generate mapping from 50Hz to 2000Hz in steps of 50Hz
audio_frequencies = np.arange(50, 2050, 50)  # 50Hz to 2000Hz in 50Hz steps

for audio_freq in audio_frequencies:
    original_freq = audio_freq * time_compression_ratio
    print(f"{audio_freq:>20.0f} | {original_freq:>25.6f}")

print(f"=" * 60)

# Add some context about ion-scale waves
print(f"\nüåä Ion-Scale Wave Context:")
print(f"   Proton gyrofrequency at 1 AU: ~1 Hz")
print(f"   Proton gyrofrequency closer to Sun: higher (scales with B-field)")
print(f"   Wave frequencies in paper: 0.41-3.03 Hz (appendix event)")

# Calculate what frequencies in the audio correspond to typical ion-scale waves
typical_wave_freq_range = [0.41, 3.03]  # From the paper
audio_equiv_range = [freq / time_compression_ratio for freq in typical_wave_freq_range]

print(f"\nüéµ Audio Equivalent of Ion-Scale Waves:")
print(f"   Original wave range: {typical_wave_freq_range[0]:.2f} - {typical_wave_freq_range[1]:.2f} Hz")
print(f"   Audio equivalent: {audio_equiv_range[0]:.0f} - {audio_equiv_range[1]:.0f} Hz")
print(f"   Listen for enhanced activity in this audio frequency range!")
```
