# ⏰ Time Clipping Behavior: What Gets Clipped and When

This notebook explains the **critical** time-clipping behavior in Plotbot that can cause unexpected results if not understood.

## 🔑 Key Concept:

**Only variables PASSED to `plotbot()` get time-clipped!**

If you access other components from the same dataset that were NOT passed to plotbot, they will return **ALL accumulated data** from the cubby, not time-clipped data.
