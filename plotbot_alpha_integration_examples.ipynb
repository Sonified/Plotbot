# PSP Alpha Particle Integration Examples

This notebook demonstrates the successful integration of PSP alpha particle data (`spi_sf0a_l3_mom`) into Plotbot.

## Key Features:
- Alpha particle moments (density, temperature, velocity)
- Direct comparison with proton data
- Temperature anisotropy analysis
- Alpha-proton drift velocities

## Physics Highlights:
- Alphas have lower density than protons (typically ~1-5% of proton density)
- Alphas are significantly hotter than protons (often 5-20x higher temperature)
- Alpha-proton velocity drift is clearly visible
- Temperature anisotropy shows different behavior between species


````python
# Import Plotbot and set up environment
import plotbot
from plotbot import *
import numpy as np
from datetime import datetime

print("ðŸš€ Plotbot Alpha Integration Demo")
print(f"Plotbot Version: {plotbot.__version__ if hasattr(plotbot, '__version__') else 'Development'}")
````

## Example 1: Alpha vs Proton Density and Temperature Comparison

This example shows the fundamental differences between alpha particles and protons in the solar wind.


````python
trange = ['2023-09-28/06:00:00.000', '2023-09-28/07:00:00.000']
````

````python
# Plot 1: Density comparison
plotbot(trange,
        psp_alpha.density, 1,
        proton.density, 2)

print("ðŸ“Š Alpha density (orange) vs Proton density (blue)")
print("Notice: Alpha density is typically 1-5% of proton density")
````

````python
# Plot 2: Temperature comparison
plotbot(trange,
        psp_alpha.temperature, 1,
        proton.temperature, 2)

print("ðŸŒ¡ï¸ Alpha temperature (red) vs Proton temperature (magenta)")
print("Notice: Alpha temperature is typically 5-20x higher than protons")
````

## Example 2: Velocity Analysis - Alpha-Proton Drift

Alpha particles typically flow faster than protons in the solar wind, creating a drift velocity that's important for plasma physics.


````python
# Plot 3: Radial velocity comparison
plotbot(trange,
        psp_alpha.vr, 1,
        proton.vr, 2)

print("ðŸš€ Alpha radial velocity (red) vs Proton radial velocity (red)")
print("Notice: Alpha particles flow faster, showing alpha-proton drift")
````

````python
# Calculate and display alpha-proton drift velocity
alpha_vr_data = psp_alpha.vr.data
proton_vr_data = proton.vr.data
drift_velocity = alpha_vr_data - proton_vr_data

print(f"ðŸ“ˆ Alpha-Proton Drift Statistics:")
print(f"   Mean drift: {np.nanmean(drift_velocity):.1f} km/s")
print(f"   Std drift:  {np.nanstd(drift_velocity):.1f} km/s")
print(f"   Range:      {np.nanmin(drift_velocity):.1f} to {np.nanmax(drift_velocity):.1f} km/s")
````

## Example 3: Temperature Anisotropy Analysis

Temperature anisotropy (TâŠ¥/Tâˆ¥) reveals important information about plasma heating and magnetic field interactions.


````python
# Plot 4: Temperature anisotropy comparison
plotbot(trange,
        psp_alpha.anisotropy, 1,
        proton.anisotropy, 2)

print("ðŸ”„ Alpha anisotropy (orange) vs Proton anisotropy (green)")
print("Notice: Different anisotropy behavior between species")
````

## Example 4: Multi-Panel Overview Plot

A comprehensive view showing all key parameters for both species.


````python
# Plot 5: Comprehensive multi-panel comparison
plotbot(trange,
        psp_alpha.density, 1,
        proton.density, 2,
        psp_alpha.temperature, 3,
        proton.temperature, 4,
        psp_alpha.vr, 5,
        proton.vr, 6,
        psp_alpha.anisotropy, 7,
        proton.anisotropy, 8)

print("ðŸ“Š Complete overview: density, temperature, velocity, and anisotropy")
print("ðŸ”¬ Perfect for comparative plasma physics analysis")
````
